```python
students=[("Urooj",61),("Fatima",97),("laiba",64)]
def grade(marks):
    if (90<=marks<=100):
        return"A"
    elif (80<=marks<=89):
        return"B"
    elif (70<=marks<=79):
        return"C"
    elif (60<=marks<=69):
        return"D"
    else:
        return"F"
grade_A=[]
grade_B=[]
grade_C=[]
grade_D=[]
grade_F=[]
for name,marks in students:
    g=grade(marks)
    if g=="A":
        grade_A.append(name)
    elif g=="B":
        grade_B.append(name)
    elif g=="C":
        grade_C.append(name)
    elif g=="D":
        grade_D.append(name)
    else:
        grade_F.append(name)
print("Grade of A student:",grade_A)
print("Grade of B student:",grade_B)
print("Grade of C student:",grade_C)
print("Grade of D student:",grade_D)
print("Grade of F student:",grade_F)
highest = students[0]
lowest = students[0]

for name, marks in students:
    if marks > highest[1]:
        highest = (name, marks)

    if marks < lowest[1]:
        lowest = (name, marks)

print("Highest Scorer:", highest)
print("Lowest Scorer:", lowest)
total_marks = 0

for name, marks in students:
    total_marks = total_marks + marks

average = total_marks / len(students)

print("Class Average:", average)
```

```python
def calculate_average(number):
    total=0
    for num in number:
        total=total+num
        average=total/len(number)
        return average
num=[10,20,30,40,50]
result=calculate_average(num)
print ("Average:",result)
```
